```python
import numpy as np
import matplotlib.pyplot as plt

a = np.genfromtxt("data.csv", delimiter = ",", dtype=str, skip_header=1)

#  year, total data group by year, meaning job is better year by year from 2020.
year = a[0:, 0].astype(int)
uyear = np.unique(year)
for y in uyear:
    print (str(y) + " year total count" + ": " + str(year[year == y].size))

# salary_in_usd, max/min/avg salary
salary = a[6:, 6].astype(int)
print("max salary in USD: " + str(np.max(salary)))
print("min salary in USD: " + str(np.min(salary)))
print("avg salary in USD: " + str(np.mean(salary)))

# job_title, print top 10 pupular job, and plot avg salary in top 10 job
job = a[3:, 3].astype(str)
job_cat = np.unique(job)
job_dict = {}
for j in job_cat:
    job_dict[j] = job[job == j].size

topJob = []
sd = sorted(job_dict.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)
print("top 10 popular job is")
for i in range(10):
    topJob.append(sd[i][0])
print(topJob)

topJobAvgSalary = []
for j in topJob:
    f = []
    f.append(j)
    filter = np.array(f)
    avg_sal = np.mean(a[(a[:,3,None] == filter[None,:]).any(axis=1)][6:, 6].astype(int))
    topJobAvgSalary.append(avg_sal)

x = np.array(topJob)
y = np.array(topJobAvgSalary)
plt.figure(figsize=(30, 15), dpi=60)
plt.bar(x,y, width=0.3)
plt.title("avg salary for different job")
plt.show()

# company_location, top 10 popular country
c_loc = a[7:, 7].astype(str)
all_country = np.unique(c_loc)
country = {}
for c in all_country:
    country[c] = c_loc[c_loc == c].size

sd = sorted(country.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)
topCountry = []
for i in range(10):
#for i in range(len(sd)):
    topCountry.append(sd[i][0])
print(topCountry)

topCountryAvgSalary = []
for c in topCountry:
    print(c)
    f = []
    f.append(c)
    filter = np.array(f)
    topCountryAvgSalary.append(np.mean(a[(a[:,7,None] == filter[None,:]).any(axis=1)][6:, 6].astype(int)))
 
x = np.array(topCountry)
y = np.array(topCountryAvgSalary)
plt.bar(x,y)
plt.show()
plt.hist(topCountryAvgSalary, label='avery salary')
plt.show()

# experience_level
e_type = a[2:, 2].astype(str)
all_type = np.unique(e_type) 
print("there are four type of employment")
e_count = []
for e in all_type:
    f = []
    f.append(e)
    filter = np.array(f)
    e_count.append(len(a[(a[:,2,None] == filter[None,:]).any(axis=1)]))

y = np.array(e_count)
plt.pie(y, labels=all_type, autopct='%.2f%%', explode=(0, 1, 1, 1))
plt.title("employment type")
plt.show()

```
